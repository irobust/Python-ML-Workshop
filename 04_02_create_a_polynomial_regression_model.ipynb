
# 04 — Create a linear regression model

This notebook uses the California Housing dataset available from scikit-learn.


## Imports

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split

from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression

from sklearn.metrics import mean_absolute_error
```

## Load the California Housing dataset

```python

california = fetch_california_housing(as_frame=True)
X = california.data
y = california.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

X_train.shape, X_test.shape
```

## Fit a Linear Regression model

```python
degree = 3
model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
model.fit(X_train, y_train)
```

## View coefficients and intercept

```python
y_predict_train = model.predict(np.array(X_train))
y_predict_test = model.predict(np.array(X_test))

trainingdata_mae = mean_absolute_error(y_train, y_predict_train)
testingdata_mae = mean_absolute_error(y_test, y_predict_test)

trainingdata_mae, testingdata_mae
```

## Evaluate the Model

```python
# output the r-squared of the model fit - % of variation explained by the model
model.score(X_test, y_test)
```

## Generate predictions

```python
predictions = model.predict(X_test)
results = pd.DataFrame({'Actual': y_test, 'Predicted': predictions})
results.sample(n=10)
```
