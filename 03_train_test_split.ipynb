
# 03 — Perform a train test split

A goal of machine learning is to build a model that performs well on new data.  
The problem is that you might not have new data available.

In this notebook, you’ll learn how to simulate unseen data by splitting what you already have into training and testing sets.


## Imports

```python

import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
```

## Load the Titanic dataset

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

## Select features and target

```python
X = titanic[["pclass", "sex", "age"]]
y = titanic["survived"]

X,y
```

## Perform the train-test split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
X_train.head(5)
# X_train.shape, X_test.shape, y_train.shape, y_test.shape
```
