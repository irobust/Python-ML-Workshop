
# 07 — Build a Decision Tree

In this notebook, you'll create and visualize a **Decision Tree** classifier using scikit-learn's `make_moons` dataset.  
Decision Trees are interpretable and let you see how models make decisions step by step.


**Example Decision Tree - Should you wear a rain jacket?**

```
              Is it raining?
                [Root node]
                 /       \
               Yes        No
                |          \
                |        [Leaf] Do not wear jacket
                |
         Is it cold outside?
            [interior node]
            /         \
          Yes         No
           |           \
   [Leaf] Wear rain    [Leaf] Take umbrella only
           jacket
```



## Import libraries

````python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_moons
````

## Generate the dataset

````python
X, y = make_moons(n_samples=1000, noise=0.25, random_state=42)

plt.figure(figsize=(5,4))
plt.scatter(X[:,0], X[:,1], c=y, cmap=plt.cm.coolwarm)
plt.title("Make moons datasets")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
````

## Train a Decision Tree model

````python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

dtree = DecisionTreeClassifier(max_depth=4, random_state=1)
dtree.fit(X_train, y_train)
````

## Visualize the Decision Tree structure

````python
plt.figure(figsize=(14,6))
plot_tree(dtree, filled=True, feature_names=["Feature 1", "Feature 2"])
plt.title("Decision Tree Visualization")
plt.show()
````

````python
from sklearn.metrics import accuracy_score
import pandas as pd

predictions = dtree.predict(X_test)

accuracy = accuracy_score(y_test, predictions)
print(f"Accuracy = {accuracy}")

pd.DataFrame({
    "Actual": y_test,
    "Predict": predictions
})
````
